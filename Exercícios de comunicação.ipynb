```python
# Solicite o nome do usuário e o exibe
nome_usuario = input("Digite seu nome inteiro: ")
print(f"olá, {nome_usuario}! Bem vindo ao nosso sistema. ")

# Solicite a idade do usuário e converta para inteiro
idade_usuario = int(input("por favor, insira sua idade: "))

# Solicite sua renda
renda_usuario = float(input("por favor, insira sua renda: "))

# Verifique se o usuário pode pegar um empréstimo
idade = idade_usuario >= 18
renda = renda_usuario >= 1000

if idade and renda:
    print("Parabéns! Você pode pegar um empréstimo.")
else:
    print("Desculpe, você não pode pegar um empréstimo. Você é menor de idade.")
```
