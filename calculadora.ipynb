```python
import tkinter as tk

class Calculadora:
    def __init__(self, root):
        self.root = root
        self.root.title("Calculadora")

        # Tela de entrada
        self.entrada = tk.Entry(root, width=16, font=('Arial', 24), borderwidth=2, relief='ridge')
        self.entrada.grid(row=0, column=0, columnspan=4)

        # Bot√µes da calculadora
        botoes = [
            ('7', 1, 0), ('8', 1, 1), ('9', 1, 2), ('/', 1, 3),
            ('4', 2, 0), ('5', 2, 1), ('6', 2, 2), ('*', 2, 3),
            ('1', 3, 0), ('2', 3, 1), ('3', 3, 2), ('-', 3, 3),
            ('0', 4, 0), ('.', 4, 1), ('=', 4, 2), ('+', 4, 3),
        ]

        for (texto, linha, coluna) in botoes:
            if texto == '=':
                botao = tk.Button(root, text=texto, width=5, height=2, command=self.calcular)
            else:
                botao = tk.Button(root, text=texto, width=5, height=2, command=lambda t=texto: self.adicionar(t))
            botao.grid(row=linha, column=coluna)

    def adicionar(self, texto):
        atual = self.entrada.get()
        self.entrada.delete(0, tk.END)
        self.entrada.insert(tk.END, atual + texto)

    def calcular(self):
        try:
            resultado = eval(self.entrada.get())
            self.entrada.delete(0, tk.END)
            self.entrada.insert(tk.END, str(resultado))
        except Exception as e:
            self.entrada.delete(0, tk.END)
            self.entrada.insert(tk.END, "Erro")

if __name__ == "__main__":
    root = tk.Tk()
    calc = Calculadora(root)
    root.mainloop()
```
